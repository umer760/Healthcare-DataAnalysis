## DATA CLEANING AND MODIFICATION

```python
dataset.drop_duplicates(inplace=True)
```

```python
dataset["Name"]=dataset["Name"].str.lstrip()
```

```python
dataset["Name"]=dataset["Name"].str.capitalize()
```

```python
dataset[["first_Name","Last_name"]]=dataset["Name"].str.split(' ',n=1,expand=True)
dataset
```

```python
dataset["Last_name"]=dataset["Last_name"].str.capitalize()
dataset
```

```python
dataset["Full_name"]=dataset["first_Name"]+' '+dataset["Last_name"]
dataset
```

```python
dataset.drop(["first_Name","Last_name"],axis=1,inplace=True)
```

```python
dataset["Name"]=dataset["Full_name"]
```

```python
dataset.drop("Full_name",axis=1,inplace=True)
```

```python
dataset["Billing Amount"]=dataset["Billing Amount"].round(2)
```

```python
dataset["Hospital"]=dataset["Hospital"].str.replace(",","")
```

```python
dataset["Hospital"]=dataset["Hospital"].str.replace("-"," and ")
```

```python
dataset["Hospital"]=dataset["Hospital"].str.lstrip("and")
```

```python
dataset["Hospital"]=dataset["Hospital"].str.rstrip("and")
```

```python
dataset["Hospital"]=dataset["Hospital"].str.lstrip()
```

```python
def age_group(age):
    if age <=14:
        return( "Child_Age Group")
    elif 14 <age <=24:
        return("Youth_Age Group")    
    elif 24<age <=40:
        return("Adult_Age Group") 
    else:
        return("Old_Age Group")  
              
```

```python
dataset["Age Group"]=dataset["Age"].apply(age_group)
```

```python
dataset["Admission Year"]=dataset["Date of Admission"].dt.year
dataset
```

```python
dataset.rename(columns={"Admission Year":"Admitting Year"},inplace=True)
```

```python
dataset["Discharging Year"]=dataset["Discharge Date"].dt.year
dataset
```

```python
dataset["Discharging Month"]=dataset["Discharge Date"].dt.month_name()
dataset
```

```python
dataset["Hospitlize Days"]=dataset["Discharge Date"]-dataset["Date of Admission"]
dataset["Hospitlize Days"]=dataset["Hospitlize Days"].dt.days
```

## PAITENTS GENDER RATIO

```python
Patients_Gender=dataset.groupby("Gender").size().reset_index(name="Patients")
print(Patients_Gender)
```

```python
colors=['hotpink','aqua']
Patients_Gender_Graph=Patients_Gender.plot.bar(x='Gender',y='Patients',color=colors,width=0.2,rot=0)
plt.bar_label(Patients_Gender_Graph.containers[0])
plt.ylim(0,30000)
plt.title("Gender Data")
plt.ylabel("Total_Patients")
plt.show()
```

## DISEASES APPEARNCE IN DIFFERENT AGE GROUPS

```python
Medical_Condition=dataset.groupby(["Age Group","Medical Condition"]).size().reset_index(name="Patients").sort_values(by="Patients",ascending=False)
Medical_Condition
```

```python
Old_Age=Medical_Condition[Medical_Condition["Age Group"]=="Old_Age Group"]
adult_age=Medical_Condition[Medical_Condition["Age Group"]=="Adult_Age Group"]
youth_age=Medical_Condition[Medical_Condition["Age Group"]=="Youth_Age Group"]
child_age=Medical_Condition[Medical_Condition["Age Group"]=="Child_Age Group"]
```

```python
fig,(axes1,axes2,axes3,axes4)=plt.subplots(1,4,figsize=(27,5))
axes1.pie(Old_Age["Patients"],labels=Old_Age["Medical Condition"],autopct='%1.1f%%')
axes2.pie(adult_age["Patients"],labels=adult_age["Medical Condition"],autopct='%1.1f%%')
axes3.bar(youth_age["Medical Condition"],youth_age["Patients"])
axes4.bar(child_age["Medical Condition"],child_age["Patients"])
#bar_label values visible
axes3.bar_label(axes3.containers[0])
axes4.bar_label(axes4.containers[0])
#XLabel adjustment with the help of rotation
axes3.set_xticklabels(youth_age["Medical Condition"],rotation=75)
axes4.set_xticklabels(child_age["Medical Condition"],rotation=75)
#TITLE
axes1.set_title("OLD AGE GROUP ")
axes2.set_title("ADULT AGE GROUP ")
axes3.set_title("YOUTH AGE GROUP ")
axes4.set_title("CHILD AGE GROUP ")
plt.show()

```

## TESTS RESULTS ACCORDING TO AGE GROUP

```python
Test_Results=dataset.groupby(["Age Group","Test Results"]).size().reset_index(name="Results").sort_values(by="Results",ascending=False)
Test_Results["Results %"]=round((Test_Results["Results"]/Test_Results["Results"].sum())*100,2)
Test_Results
```

```python
Old_Age=Test_Results[Test_Results["Age Group"]=="Old_Age Group"]
adult_age=Test_Results[Test_Results["Age Group"]=="Adult_Age Group"]
youth_age=Test_Results[Test_Results["Age Group"]=="Youth_Age Group"]
child_age=Test_Results[Test_Results["Age Group"]=="Child_Age Group"]
```

```python
fig,(axes1,axes2,axes3,axes4)=plt.subplots(1,4,figsize=(27,5))
axes1.pie(Old_Age["Results"],labels=Old_Age["Test Results"],autopct='%1.1f%%')
axes2.pie(adult_age["Results"],labels=adult_age["Test Results"],autopct='%1.1f%%')
axes3.bar(youth_age["Test Results"],youth_age["Results"])
axes4.bar(child_age["Test Results"],child_age["Results"])
#bar_label values visible
axes3.bar_label(axes3.containers[0])
axes4.bar_label(axes4.containers[0])
for tick in axes3.get_xticklabels():
    tick.set_fontweight('bold')
for tick in axes4.get_xticklabels():
    tick.set_fontweight('bold')
    
for tick in axes3.get_yticklabels():
    tick.set_fontweight('bold')
for tick in axes4.get_yticklabels():
    tick.set_fontweight('bold')

axes1.set_title("OLD AGE GROUP ")
axes2.set_title("ADULT AGE GROUP ")
axes3.set_title("YOUTH AGE GROUP ")
axes4.set_title("CHILD AGE GROUP ")
plt.show()
```
